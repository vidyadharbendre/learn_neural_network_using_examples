### Tensor name and shape

```
hello_2 =  tf.constant('Hello World', shape=[6,2])
print(hello_2.numpy())
```

### Adding two tensors

```
a = tf.constant([1.2])
b = tf.constant([3.4])
c = tf.add(a,b)
print(c.numpy())
```

```
a = 1.2
b = 3.4
c = tf.add(a,b)
print(c)
```
