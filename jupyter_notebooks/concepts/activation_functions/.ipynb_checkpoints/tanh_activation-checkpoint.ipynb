```python
import numpy as np
import matplotlib.pyplot as plt

# Hyperbolic tangent function
def tanh(z):
    return np.tanh(z)

# Derivative of tanh function
def tanh_derivative(z):
    return 1 - np.tanh(z)**2

# Function to plot tanh and its derivative
def plot_tanh_and_derivative(start, end, step):
    z = np.arange(start, end, step)
    y_tanh = tanh(z)
    y_derivative = tanh_derivative(z)

    plt.figure(figsize=(10, 6))
    plt.plot(z, y_tanh, label='Tanh Function')
    plt.plot(z, y_derivative, label='Derivative of Tanh Function')
    plt.xlabel('z')
    plt.ylabel('Value')
    plt.title('Tanh Function and Its Derivative')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot tanh and its derivative for the range -2 to 2
plot_tanh_and_derivative(-2, 2, 0.1)
```

```python
# Plot tanh and its derivative for the range -2 to 2
plot_tanh_and_derivative(-4, 4, 0.1)
```
