```python
import numpy as np
import matplotlib.pyplot as plt 
```

# Sigmoid Function and Its Derivative Plotter

This Python script plots the sigmoid function and its derivative for given ranges of input values.

## Description

The sigmoid function, also known as the logistic function, is commonly used in machine learning and neural networks. It maps any real-valued number to the range (0, 1). The derivative of the sigmoid function is used in gradient descent algorithms for training neural networks.

This script defines functions to calculate the sigmoid function and its derivative, and then plots both functions for specified ranges of input values.


```python
# def sigmoid(z):
#     return 1 / (1 + np.exp(-z))
```

```python
def sigmoid(z):
    return np.exp(z) / (1 + np.exp(z))
```

```python
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

```python
def plot_sigmoid_and_derivative(start, end, step):
    z = np.arange(start, end, step)
    y_sigmoid = sigmoid(z)
    y_derivative = sigmoid_derivative(z)

    plt.figure(figsize=(10, 6))
    plt.plot(z, y_sigmoid, label='Sigmoid Function')
    plt.plot(z, y_derivative, label='Derivative of Sigmoid Function')
    plt.xlabel('z')
    plt.ylabel('Value')
    plt.title('Sigmoid Function and Its Derivative')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Plot sigmoid and its derivative for the range -5 to 5
plot_sigmoid_and_derivative(-5, 5, 0.1)
```

```python
# Plot sigmoid and its derivative for the range -100 to 100
plot_sigmoid_and_derivative(-10, 10, 1)
```
