```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU (Rectified Linear Unit) function
def relu(z):
    return np.maximum(0, z)

# Derivative of ReLU function
def relu_derivative(z):
    return np.where(z > 0, 1, 0)

# Function to plot ReLU and its derivative
def plot_relu_and_derivative(start, end, step):
    z = np.arange(start, end, step)
    y_relu = relu(z)
    y_derivative = relu_derivative(z)

    plt.figure(figsize=(10, 6))
    plt.plot(z, y_relu, label='ReLU Function')
    plt.plot(z, y_derivative, label='Derivative of ReLU Function')
    plt.xlabel('z')
    plt.ylabel('Value')
    plt.title('ReLU Function and Its Derivative')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot ReLU and its derivative for the range -2 to 2
plot_relu_and_derivative(-2, 2, 0.1)
```

```python
# Plot ReLU and its derivative for the range -2 to 2
plot_relu_and_derivative(-3, 3, 0.1)
```

```python
# Plot ReLU and its derivative for the range -2 to 2
plot_relu_and_derivative(-1, 1, 0.1)
```
