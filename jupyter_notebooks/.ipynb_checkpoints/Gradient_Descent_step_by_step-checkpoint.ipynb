# Gradient Descent

```python
#This example is from the appendix section of the faculty deck
import numpy as np

Y = np.array([-61.53,-19.19,401.13,601.19,799.28,584.84,1156.96,901.36,1126.31,1280.4])
x = np.array([  1,  21,  41,  61,  81, 101, 121, 141, 161, 181])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
#plt.plot(x , Y , color = 'red', label = 'actual')
plt.scatter(x , Y , color = 'red', label = 'actual')
plt.legend()
plt.show()
```

```python
N = len(x)
```

```python
#Function which we want to optimize (a, b are the weights; x is the independent variable)
def func(a,b): 
    f = a + b * x
    return f
```

```python
# Loss function which is actual value minus predicted value; Here, Y is the actual value
def loss(a,b): #
    lo = Y - func(a,b)
    return lo
```

```python
# Cost function - Mean Square Error 
def Cost(a,b):
    cost = (1/2)*(((loss(a,b))**2)/N).sum()
    return cost
```

```python
# Derivative of the above cost function with respect to a
def Cost_der_a(a,b,x,Y):   
    return -sum((1/N) *(loss(a,b)))

# Derivative of the above cost function with respect to b
def Cost_der_b(a,b,x,Y):   
    return -(1/N) * ((x.dot(loss(a,b))))
```

```python
#Set up some initial values for the weights
a_in = 0 
b_in = 0
learning_rate = .001
error_rate = .001  #Threshold error rate
list_a = [a_in]
list_b = [b_in]
i = 0
error_a , error_b = .1 , .1
error_a_list, error_b_list = [error_a] , [error_b]

while (np.abs(error_a) > error_rate) or (np.abs(error_b) > error_rate): # Iterate until the error is below the threshold
    # Calculating gradients
    A = Cost_der_a(list_a[i],list_b[i],x,Y)
    B = Cost_der_b(list_a[i],list_b[i],x,Y) 
    
    #Calculating new value of weights a,b and taking a step(based on learning rate) towards the direction of gradient.
    a_in -=  (learning_rate * A / N)  
    b_in -=  (learning_rate * B / N) 
    
    list_a.append(a_in)
    list_b.append(b_in)
    error_a = list_a[i] - list_a[i+1]
    error_b = list_b[i] - list_b[i+1]
    i += 1
    error_a_list.append(error_a)
    error_b_list.append(error_b) 
    #print("error_a",error_a ,"error_b" ,error_b ,'a_in',a_in,'b_in',b_in, "Cost", Cost(a_in,b_in))
    continue
print('Final values of weights: ', a_in , b_in)
```

```python
#Plot showing evolution of the best fit line. It plots the function for every iteration of gradient descent
fig, axs = plt.subplots(2,4, figsize=(22,10))
for i in range(len(list_a)):
    plt.subplot(2,4,i+1)
    plt.plot(x , (list_a[i] + list_b[i]*x) , color = 'blue', label = 'predicted')
    plt.scatter(x , Y , color = 'red', label = 'actual')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Iteration {0}: a={1:.2f}, b={2:.2f}'.format(i, list_a[i], list_b[i]))
    plt.legend()
plt.show()
```

```python
#Plot to show the errors after each iteration:
fig,axs = plt.subplots(nrows=1, ncols=2,figsize=(12,4))
plt.subplot(1,2,1)
plt.plot([p for p in range(1,len(error_a_list)+1)], np.abs(error_a_list), color = 'orange')
plt.xlabel('Number of Iterations')
plt.ylabel('Error in a')
plt.subplot(1,2,2)
plt.plot([x for x in range(1,len(error_b_list)+1)], np.abs(error_b_list), color = 'orange')
plt.xlabel('Number of Iterations')
plt.ylabel('Error in b')
```

```python
#Plot showing the best fit line that we predicted by optimizing weights a,b using gradient descent
plt.figure(figsize=(10,5))
plt.plot(x , (a_in + b_in*x) , color = 'blue', label = 'predicted')
plt.scatter(x , Y , color = 'red', label = 'actual')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
