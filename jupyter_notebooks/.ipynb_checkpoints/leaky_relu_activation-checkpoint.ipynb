```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky ReLU function
def leaky_relu(z, alpha=0.01):
    return np.where(z > 0, z, alpha * z)

# Derivative of Leaky ReLU function
def leaky_relu_derivative(z, alpha=0.01):
    return np.where(z > 0, 1, alpha)

# Function to plot Leaky ReLU and its derivative
def plot_leaky_relu_and_derivative(start, end, step):
    z = np.arange(start, end, step)
    y_leaky_relu = leaky_relu(z)
    y_derivative = leaky_relu_derivative(z)

    plt.figure(figsize=(10, 6))
    plt.plot(z, y_leaky_relu, label='Leaky ReLU Function')
    plt.plot(z, y_derivative, label='Derivative of Leaky ReLU Function')
    plt.xlabel('z')
    plt.ylabel('Value')
    plt.title('Leaky ReLU Function and Its Derivative')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Leaky ReLU and its derivative for the range -2 to 2
plot_leaky_relu_and_derivative(-2, 2, 0.1)
```

```python
# Plot Leaky ReLU and its derivative for the range -2 to 2
plot_leaky_relu_and_derivative(-5, 5, 0.1)
```

```python
# Plot Leaky ReLU and its derivative for the range -2 to 2
plot_leaky_relu_and_derivative(-100, 50, 0.1)
```
