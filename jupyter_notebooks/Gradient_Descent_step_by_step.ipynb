# Gradient Descent

Here's a breakdown of key points about gradients in machine learning:

Gradient of a Function: In mathematics, the gradient of a scalar-valued function of several variables is a vector that points in the direction of the greatest rate of increase of the function at a particular point. It is defined as the vector of partial derivatives of the function with respect to each variable.

Gradient Descent: In machine learning, gradient descent is an optimization algorithm used to minimize a loss function by iteratively updating the parameters of a model. At each iteration, the gradient of the loss function with respect to the model parameters is computed, and the parameters are adjusted in the direction opposite to the gradient to reduce the loss.

Learning Rate: The size of the steps taken in the direction of the gradient during gradient descent is determined by the learning rate hyperparameter. A larger learning rate results in larger steps, which can lead to faster convergence but may overshoot the minimum. A smaller learning rate may take longer to converge but is less likely to overshoot.

Local Minima and Global Minima: Gradient descent is not guaranteed to find the global minimum of a function, as it can get stuck in local minima or saddle points. However, for many convex and smooth loss functions encountered in machine learning, gradient descent often converges to a good solution.

Gradient in Deep Learning: In deep learning, gradients are used extensively in training neural networks. During the training process, the gradients of the loss function with respect to the weights of the network are computed using techniques such as backpropagation. These gradients are then used to update the weights of the network in order to minimize the loss and improve the model's performance on the training data.

In summary, the gradient plays a crucial role in optimization algorithms like gradient descent, enabling the training of machine learning models by iteratively adjusting their parameters to minimize a given loss function

```python
import numpy as np
import matplotlib.pyplot as plt

def sse_loss(y_true, y_pred):
    """
    Compute the Sum of Squared Errors (SSE) loss.

    Parameters:
        y_true: True labels (array)
        y_pred: Predicted labels (array)

    Returns:
        SSE loss value.
    """
    return np.sum((y_true - y_pred) ** 2)

def gradient_descent_step(x, y, w, b, learning_rate):
    """
    Perform one step of gradient descent for linear regression.

    Parameters:
        x: Input feature (array)
        y: True labels (array)
        w: Weight parameter
        b: Bias parameter
        learning_rate: Learning rate for gradient descent

    Returns:
        Updated weight and bias parameters after one step of gradient descent.
    """
    # Compute predictions
    y_pred = w * x + b
    
    # Compute gradients
    w_gradient = -2 * np.mean(x * (y - y_pred))  # Partial derivative with respect to w
    b_gradient = -2 * np.mean(y - y_pred)       # Partial derivative with respect to b
    
    # Update parameters
    w -= learning_rate * w_gradient
    b -= learning_rate * b_gradient
    
    return w, b

# Given X and Y values
Y = np.array([50, 70, 60, 78, 80, 70, 90, 86, 95, 70])
X = np.array([40, 45, 50, 60, 65, 70, 80, 75, 90, 95])

# Randomly initialize parameters
initial_w = np.random.randn()
initial_b = np.random.randn()
learning_rate = 0.0001

# Plot initial data points
plt.scatter(X, Y, label='Data Points')

# Perform gradient descent for 5 iterations
for i in range(5):
    # Perform one step of gradient descent
    initial_w, initial_b = gradient_descent_step(X, Y, initial_w, initial_b, learning_rate)
    
    # Compute predictions
    y_pred = initial_w * X + initial_b
    
    # Compute SSE loss
    loss = sse_loss(Y, y_pred)
    
    # Plot the fit line
    fit_line = initial_w * X + initial_b
    plt.plot(X, fit_line, label=f'Iteration {i+1}, Loss: {loss:.2f}')
    
    # Display w and b values
    plt.text(X[0], Y[-1] - i * 10, f'w = {initial_w:.2f}, b = {initial_b:.2f}', fontsize=8)

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gradient Descent with SSE Loss')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Given X and Y values
Y = np.array([50, 70, 60, 78, 80, 70, 90, 86, 95, 70])
X = np.array([40, 45, 50, 60, 65, 70, 80, 75, 90, 95])

# Add constant term for intercept
X = sm.add_constant(X)

# Fit linear regression model
model = sm.OLS(Y, X).fit()

# Get predicted values
Y_pred = model.predict(X)

# Plot data points
plt.scatter(X[:, 1], Y, label='Data Points')

# Plot fit line
plt.plot(X[:, 1], Y_pred, color='red', label='Fit Line')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression with statsmodels')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
